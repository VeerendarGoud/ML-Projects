```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# functional methond

plt.plot(x,y,'r-')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.title('x, y',fontsize=16)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')

plt.show()
```

```python
# OO

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('x lable')
axes.set_ylabel('y lable')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('large')


axes2.plot(y,x)
axes2.set_title('small')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)


axes[0][0].plot(x,y)
axes[0][0].set_title('one')


axes[1][1].plot(x,y)
axes[1][1].set_title('two')

#axes.plot(x,y)

#for current_ax in axes:
#    current_ax.plot(x,y)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0)
```

```python
fig,axes = plt.subplots(figsize=(8,5),nrows=2,ncols=1)

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.plot(x,y)

ax1 = fig.add_axes([0.1,0.6,0.4,0.3])
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax2 = fig.add_axes([.2,.5,.4,.4])
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.plot(x,y)
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2,figsize=(10,5))

axes[0].plot(x,y,color='blue',ls='--',lw=2)
axes[1].plot(x,z,color='red',ls='-',lw=2)
```
